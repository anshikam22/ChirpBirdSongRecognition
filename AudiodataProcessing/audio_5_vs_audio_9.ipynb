**Conclusion  Audio signal of 5 vs 9**         


*   We can see that the pitch of audio signal 5 is continuous while that of 9 breaks at one point.
*   Also pitch of audio signal 5 is higher than that of 9.


*   We can also see that the zero crossing rate of 5 ranges from 0.02 to 0.19 while that of 9 ranges from 0.01 to 0.09.
*   Also the root mean square energy of audio signal of 5 has a much higher range than that of ausio signal of 9.


*  Onset strength of audio signal 5 is around 15 almots double of onset strength of 9 .
*   The range of rolloff frequencies attained by the frames is much much higher(around 4000 hertz) in audio signal of 5 than around 700 Hertz in audio signal of 9.


*   The amplitude range of audio '5' is around -0.10 to 0.10 which larger than that of audio '9' (-0.06 t0 0.02)
*   Looking closely at the spectograms, we gather that there are more lighter parts in audio signal of 9 than in 5.
