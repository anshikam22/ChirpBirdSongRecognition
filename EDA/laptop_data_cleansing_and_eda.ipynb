```python
def x(record):
  if record in ['Intel Core i7', 'Intel Core i5','Intel Core i3']:
    return record
  else:
        if record.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
df['Cpu_brand']=df['Cpu_name'].apply(x)
```

```python
def naming(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Chrome OS/ Android /No OS/Linux'
df['os'] = df['OpSys'].apply(naming)
#making another column os by cleaning OpSys column.
```
